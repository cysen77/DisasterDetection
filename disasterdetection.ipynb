```python
#网络模型
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.bert = AutoModel.from_pretrained("bert-base-uncased")
        self.predict=nn.Sequential(
            nn.Linear(768,256),
            nn.ReLU(),
            nn.Linear(256,2),
            nn.Softmax())

    def forward(self,x):
        x = self.bert(**x).last_hidden_state[:, 0, :]
        x=self.predict(x)
        return x
```

```python
#读取数据并划分训练集和验证集
pd_train_set=pd.read_csv("/kaggle/input/nlp-getting-started/train.csv")
pd_train_set.text = pd_train_set.text.apply(clean)
pd_train_set=pd_train_set[["text","target"]]
#训练集
validation_set=pd_train_set.sample(frac=line)
#验证集
train_set=pd_train_set[~pd_train_set.index.isin(validation_set.index)]
```

```python
#构造dataset
class MyDataset(Dataset):
    def __init__(self,cls="train"):

        if cls=="train":
            self.data=train_set
        elif cls=="validation":
            self.data=validation_set
        else:
            raise Exception("unknown name")

    def __getitem__(self, index):
        data=self.data.iloc[index]
        text,target=data
        return text,target

    def __len__(self):
        return len(self.data)
```

```python
#自定义DataLoader中的collate_fn参数
#在加载dataset的时候，将其中的文本数据转化为Tensor张量
#最后的返回结果是一个元组(texts,targets),texts为batch_size个文本张量，targets为batch_size个标签
def my_collate(batch):
    texts=[text for text,target in batch]
    targets=[target for text,target in batch]
    #调用预训练好的分词器对文本进行向量化，将每一个文本句子转换成一个长度为max_length的向量，长的截断，少的补齐
    texts=tokenizer(texts,padding="max_length",truncation=True,max_length=text_max_length,return_tensors="pt")
    targets=torch.tensor(targets,dtype=torch.long)
    return texts.to(device),targets.to(device)
```

```python
#构造训练集和验证集的dataloader
train_data=MyDataset("train")
validation_data=MyDataset("validation")
train_loader=DataLoader(dataset=train_data,batch_size=batch_size,shuffle=True,drop_last=True,collate_fn=my_collate)
validation_loader=DataLoader(dataset=validation_data,batch_size=batch_size,drop_last=True,collate_fn=my_collate)
```

```python
#初始化网络模型
net=Net()
net.to(device)
#损失函数
loss_fn=torch.nn.CrossEntropyLoss()
#优化器
optim=torch.optim.Adam(net.parameters(),lr=learning_rate)
```

```python
#训练函数
def train(loader):
    sum_loss=0
    i=0
    for data in loader:
        input,target=data
        output=net(input)
        loss=loss_fn(output,target)
        optim.zero_grad()
        loss.backward()
        optim.step()
        sum_loss+=loss.item()
        #print("第{}组数据训练完成".format(i+1))
        i+=1
    return sum_loss/len(loader)
```

```python
#验证函数
def validate(loader):
    sum_loss=0
    correct=0
    N=len(loader)*batch_size
    prediction = []
    with torch.no_grad():
        for data in loader:
            input,target=data
            output=net(input)
            loss=loss_fn(output,target)
            sum_loss+=loss.item()
            correct += (torch.argmax(output,1)==target).sum().detach().cpu().numpy()
            
    accuracy=correct/N
    #返回模型在验证集上的损失值和正确率
    return sum_loss/len(loader),accuracy
```

```python
#收集模型在验证集上的损失值和正确率，便于后续作图
list_loss=[]
list_rate=[]
train_losshistory = []
x=list(range(epoch))
#记录下验证集上最好的正确率，保存模型
best_rate=0
```

```python
#训练模型并在验证集上测试
start=time.time()
for i in range(epoch):
    trainloss=train(train_loader)
    loss,accuracy=validate(validation_loader)
    end=time.time()
    minute=int((end-start)//60)
    second=int((end-start)%60)
    print(f"轮次: {i+1:3d}/{epoch} 时间: {minute:3d}:{second:02d} 训练损失: {trainloss:.8f} 验证损失: {loss:.8f} 验证准确率: {accuracy*100:.2f}%")

    list_loss.append(loss)
    list_rate.append(accuracy)
    train_losshistory.append(trainloss)
```

```python
plt.figure(figsize=(6,4))
plt.plot(list(range(len(train_losshistory))),train_losshistory)
plt.plot(list(range(len(list_loss))),list_loss)
plt.plot(list(range(len(list_rate))),list_rate)
plt.legend(['training loss','validation loss','validation score'])
plt.xlabel('epoch')
plt.show()
```
